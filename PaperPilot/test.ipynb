```python
from pdfminer.high_level import extract_text

# Path to the PDF file you want to extract text from
pdf_file = "AR\\106\\2886_兆豐金.pdf"

# Define the page range you want to extract (e.g., pages 1 to 3)
start_page = 1
end_page = 3

# Extract text from the specified page range
text = extract_text(pdf_file, page_numbers=range(start_page, end_page + 1))

print(text)
```

```python
import fitz  # PyMuPDF

def pdf_to_xml(pdf_path, xml_path):
    try:
        pdf_document = fitz.open(pdf_path)
        xml_content = '<?xml version="1.0" encoding="UTF-8" ?>\n<document>\n'
        
        for page_num in range(pdf_document.page_count):
            page = pdf_document.load_page(page_num)
            xml_content += f'<page number="{page_num + 1}">\n'
            xml_content += page.get_text("xml")
            xml_content += '</page>\n'
        
        xml_content += '</document>'
        
        # Write the XML content to a file
        with open(xml_path, 'w', encoding='utf-8') as xml_file:
            xml_file.write(xml_content)
        
        pdf_document.close()
        print(f"PDF successfully converted to XML: {xml_path}")
        
    except Exception as e:
        print(f"Error: {e}")

# Usage
input_pdf_path = "AR\\106\\1460_宏遠.pdf"  # Replace with your PDF file path
output_xml_path = "output.xml"  # Path for the resulting XML file

pdf_to_xml(input_pdf_path, output_xml_path)
```
